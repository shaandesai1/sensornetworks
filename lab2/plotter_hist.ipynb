```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
import pandas as pd
```

```python
f = pd.read_csv('orig_vals_parsed.txt',delimiter=' ',header=None)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(10,10))
sns.distplot(f[0],norm_hist=False)
fs = 20
plt.title('Original XYZ Error Histogram',fontsize=fs)
plt.xlabel('Error_XYZ (m)',fontsize=fs)
plt.ylabel('Frequency',fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.savefig('origxyz.svg',bbox_inches='tight')
```

```python
plt.figure(figsize=(10,10))
sns.distplot(f[1],norm_hist=False)
fs = 20
plt.title('Original Angle Error Histogram',fontsize=fs)
plt.xlabel('Angular Error ($^\circ$)',fontsize=fs)
plt.ylabel('Frequency',fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.savefig('origangle.svg',bbox_inches='tight')
```

```python
plt.figure(figsize=(15,10))
fs = 20
plt.scatter(range(len(f[0])),f[0],s=0.5,alpha=0.8)
plt.xlabel('Image-index',fontsize=fs)
plt.ylabel('Error (XYZ)',fontsize=fs)
plt.xticks(fontsize=fs)
plt.yticks(fontsize=fs)
plt.title("Visual Odometry Error",fontsize=fs)
plt.savefig('viconerr.svg',bbox_inches='tight')
```
