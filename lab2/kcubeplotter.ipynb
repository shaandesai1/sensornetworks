```python
%matplotlib inline

import os
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
from matplotlib import cm
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

df = pd.read_csv('./2019group1.csv')

# print(len(df))
df.columns = ['time','field.header.seq','field.header.stamp','frame_id','child_frame_id','x','y','z',
             'x_rot','y_rot','z_rot','w_rot']

df['rotation_mag'] = df[['x_rot','y_rot','z_rot']].apply(lambda x: np.sqrt(x[0]**2 + x[1]**2 +x[2]**2),axis=1)
# df['velocity'] = df[['x_rot','y_rot','z_rot']]
# print(df.head())
ax.plot(np.array(df['y']),np.array(df['x']),np.array(df['z']))

plt.title('KaurCuup\u2122')
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.tight_layout()
plt.savefig('./KaurCuup.pdf')
plt.show()
```
